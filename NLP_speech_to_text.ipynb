```python
import pandas as pd
import re

from nltk.corpus import stopwords
import nltk

from sklearn.feature_extraction.text import CountVectorizer


from sklearn.model_selection import train_test_split

```

```python
df = pd.read_csv('/content/Mario-Gamal-Sobhy-NLP-with-speech-to-text-machine-learning/emotions.txt', sep=';',names=["Text","Emotion"])
df.sample(5)
```

```python
df['Emotion_num'] = df.Emotion.map({
    'joy' : 0,
    'sadness': 1,
    'anger': 2,
    'fear': 3,
    'love': 4,
    'surprise':5
})
```

```python
df.head(20)
```

```python
!pip install nltk
nltk.download('stopwords')
```

```python
def preprocessing(Text):
 corpus = []

 for i in range(len(df)):
     s = re.sub("[^A-Za-z]", " ", Text[i])  #remove Punctuation
     s = s.lower() # beacause all stop words is lower
     s = s.split() # Tokenize the text into words (list)
     s = [word for word in s if word not in stopwords.words("english")] # remove Stop Words
     s = " , ".join(s) # convert list into string

     corpus.append(s)

 return corpus
```

```python
corpus = preprocessing(df["Text"])
corpus
```

```python
y = df['Emotion_num']
pd.DataFrame(y).head(10)
```

```python
from sklearn.model_selection import train_test_split

# extract the labels from the train data
# use 70% for the training and 30% for the test
x_train, x_test, y_train, y_test = train_test_split(corpus, y,
                                                    stratify=y, #stratify=y helps prevent potential biases
                                                    random_state=42,
                                                    test_size=0.3, shuffle=True)
```

```python


vectorizer = CountVectorizer(binary=True, stop_words='english')
vectorizer.fit(list(x_train) + list(x_test))
# transform documents to document-term matrix
x_train_vec = vectorizer.transform(x_train)
x_test_vec = vectorizer.transform(x_test)

```

# **speech to text**

```python
#text = str(input())
import speech_recognition as sr
import pyaudio
import wave

frames_per_buffer= 3200  # Record in buffer of 3200 samples
format= pyaudio.paInt32 #32 bit per sample
channels= 1    #mono channel
sample_rate= 16000

p= pyaudio.PyAudio() #pyaudio object
stream= p.open(format= format,
               channels= channels,
               rate= sample_rate,
               input= True, #to capture audio
               frames_per_buffer= frames_per_buffer)

print("Start Recording")

seconds=5 #record duration
frames=[] # Initialize array to store frames

# Store data in buffer for 3 seconds
for i in range(0, int(sample_rate / frames_per_buffer * seconds)):
    data = stream.read(frames_per_buffer)#read 3200 frame at each iteration
    frames.append(data)

print("Finished Recording")

stream.stop_stream()
stream.close()
# Terminate the PortAudio interface
p.terminate()


wf= wave.open("output.wave", "wb")
wf.setnchannels(channels)
wf.setsampwidth(p.get_sample_size(format))
wf.setframerate(sample_rate)
#write all frames in binary string, combine all frames into binary string .
wf.writeframes(b''.join(frames))
wf.close()


audio_file = sr.AudioFile("output.wave")
#print(type(audio_file))
recognizer= sr.Recognizer()
recognizer.energy_threshold=300

with audio_file as source:
    #for noisy audio files
    recognizer.adjust_for_ambient_noise(source, duration=0.5)
    audio_file_data= recognizer.record(source)
print(type(audio_file_data))
text= recognizer.recognize_google(audio_data=audio_file_data, language="en-US")
print(text)

testing(text)

```
